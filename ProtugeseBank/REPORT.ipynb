
# 🏦 Project Report: Portuguese Bank Marketing Prediction



## 1. 📝 Problem Statement
The Portuguese Bank conducts direct marketing campaigns (phone calls) to sell term deposits. The current strategy of calling customers randomly is inefficient and has a low success rate.
**The Goal:** Build a Machine Learning model to predict which customers will say "Yes" to a term deposit, allowing the bank to target only high-potential leads.

---

## 2. 🛠️ Methodology
We followed a structured Data Science workflow:

### **A. Data Cleaning & EDA**
* **Dataset:** 41,188 rows, 21 columns.
* **Imbalance:** The target variable `y` is highly imbalanced (89% 'No' vs 11% 'Yes').
* **Data Leakage Prevention:** We dropped the `duration` column because the call duration is not known before making the call. Including it would create an unrealistic model.
* **Insight:** Economic indicators like `euribor3m` (Interest Rate) and `nr.employed` (Employment Rate) showed strong correlations with subscription success.

### **B. Preprocessing**
* **Categorical Encoding:** Used **One-Hot Encoding** for variables like 'Job', 'Marital Status', and 'Education'.
* **Scaling:** Applied **StandardScaler** to numerical features.
* **Splitting:** Used **Stratified Sampling** (70/30 split) to ensure the 'Yes' class was equally represented in training and testing.

### **C. Model Building**
We trained and evaluated 6 different algorithms:
1. **Logistic Regression:** (Baseline) - Fast but struggled with non-linear patterns.
2. **Random Forest:** Good performance, capturing complex customer behaviors.
3. **Gradient Boosting:** Strong performer.
4. **XGBoost:** **The Champion Model.** Best balance of speed and accuracy.
5. **KNN:** Moderate performance; struggled with high dimensionality.
6. **SVM:** Computationally expensive and slow for this dataset size.

# Task 1: Complete Data Analysis Report

### Dataset Overview

The dataset consists of direct marketing campaigns (phone calls) of a Portuguese banking institution. The goal was to analyze customer behavior and predict if they would subscribe to a term deposit. The dataset contains 41,188 rows and 21 columns, including client demographics, campaign history, and economic indicators.

### Data Cleaning and Integrity

The first step was to ensure data quality. We checked for null values and identified that missing data was labeled as "unknown." We decided to treat "unknown" as a valid category rather than imputing it, as the lack of information can sometimes be predictive. We removed duplicate rows to ensure statistical validity.

### Critical Adjustment: Removing Data Leakage

A crucial step in the analysis was identifying the "duration" column (length of the call). This variable is not known before a call is performed. Including it would cause data leakage, creating an unrealistic model that cheats by knowing the future. Therefore, the "duration" column was dropped entirely from the dataset.

### Exploratory Data Analysis (EDA) Findings

1. Target Imbalance: The dataset is highly imbalanced, with approximately 89% of clients refusing the deposit (No) and only 11% subscribing (Yes).
2. Seasonality: The month of May sees the highest call volume but a very low conversion rate. Months like March, September, October, and December have lower volume but significantly higher success rates.
3. Economic Impact: There is a strong negative correlation between the Euribor 3-month interest rate and subscriptions. When interest rates are high, subscriptions drop.
4. Demographics: Students and Retired individuals have higher subscription rates compared to Admin or Blue-collar workers.

# Task 2: Predictive Model Strategy

### Objective

The objective was to build a machine learning model that allows the bank marketing team to prioritize which customers to call, rather than calling randomly.

### Methodology

We employed a Supervised Learning approach. The data was split into a Training Set (70%) to teach the model and a Testing Set (30%) to evaluate it. We used Stratified Sampling to ensure the minority class (subscribers) was represented equally in both sets.

### Algorithm Selection

We trained multiple algorithms, including Logistic Regression, Random Forest, XGBoost, KNN, and SVM. We selected XGBoost as the final model because it handles the mix of categorical data (jobs, education) and numerical data (economic rates) effectively and provided the highest accuracy and AUC score.

### Business Value

This predictive model assigns a probability score to every customer. The bank can now filter the list and only call customers with a high probability of saying "Yes." This reduces call center costs, saves time, and improves the efficiency of the marketing campaign.

# Task 3: Suggestions to the Bank Marketing Team

Based on the insights from our best model (XGBoost), we recommend the following strategies to increase term deposit sales:

1. Capitalize on Low Interest Rates
The data shows that customers are most likely to invest when the Euribor 3-month rate is low. The marketing team should save their budget and launch aggressive call campaigns specifically during "low-rate windows" (typically below 1.5%).

2. The VIP Re-Targeting Strategy
The "poutcome" (previous campaign outcome) is the strongest predictor of success. Customers who successfully subscribed in a previous campaign are statistically 50 times more likely to buy again. These customers should be placed on a VIP Call List and contacted first.

3. Prioritize Mobile Contacts
Calls made to cellular phones have a significantly higher conversion rate than landlines. We suggest updating data collection forms to prioritize mobile numbers and potentially using SMS notifications to warm up leads before calling.

4. Seasonal Optimization
Stop focusing heavily on May, which has high volume but low success. Shift resources to October, September, March, and December. These months show higher customer engagement and conversion.

# Report on Challenges Faced

During the project, we encountered several data and technical challenges. Here is how we addressed them:

1. Challenge: Data Leakage
Issue: The dataset contained a column named "duration" (call length). This feature is highly predictive but is only known after a call ends. Using it would make the model useless for future prediction.
Technique Used: We dropped the "duration" column from the dataset before training.
Reason: To ensure the model is realistic and can predict outcomes before the call is made.

2. Challenge: Class Imbalance
Issue: The dataset was heavily skewed, with 89% of outcomes being "No" and only 11% "Yes." A standard model might just predict "No" for everyone and still be 89% accurate.
Technique Used: We used Stratified Train-Test Splitting and evaluated models using Area Under the Curve (AUC) and Confusion Matrices rather than just Accuracy.
Reason: To ensure the model actually learns to identify the minority class (subscribers).

3. Challenge: High Dimensionality
Issue: Columns like "Job" and "Education" are text-based. Converting them to numbers using One-Hot Encoding created many new columns, making the dataset complex.
Technique Used: We used the Scikit-Learn Pipeline and ColumnTransformer.
Reason: This automated the transformation process, preventing errors and keeping the code clean and efficient.

4. Challenge: Software Compatibility
Issue: We initially attempted to build an Artificial Neural Network (ANN) using TensorFlow but faced DLL/Library errors in the environment.
Technique Used: We pivoted to using XGBoost.
Reason: XGBoost is often superior to Neural Networks for tabular data, is easier to install, and provided excellent results without the technical overhead.
