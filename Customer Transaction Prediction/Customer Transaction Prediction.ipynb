### BALANCING THE TARGET COLUMN USING SMOTE

```python
from collections import Counter
from imblearn.over_sampling import SMOTE #SMOTE(synthetic minority oversampling techinque)
sm = SMOTE() # obeject creation
print("unbalanced data   :  ",Counter(y))
X_sm,y_sm = sm.fit_resample(X,y)
print("balanced data:    :",Counter(y_sm))
```

### SPLIT TRAINING AND TESTING DATA

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import RandomOverSampler
import numpy as np

# ==========================
# 1. TRAIN TEST SPLIT
# ==========================
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.30,
    random_state=42,
    stratify=y
)

print("Before Balancing:")
print("X_train shape:", X_train.shape)
print("y_train distribution:\n", y_train.value_counts())

# ==========================
# 2. MEMORY OPTIMIZATION
# ==========================
# Convert to float32 to reduce RAM by 50%
if hasattr(X_train, "astype"):
    X_train = X_train.astype(np.float32)
    X_test  = X_test.astype(np.float32)
else:
    X_train = X_train.to_numpy(dtype=np.float32)
    X_test  = X_test.to_numpy(dtype=np.float32)

# ==========================
# 3. SAFE CLASS BALANCING (NO SMOTE CRASH)
# ==========================
ros = RandomOverSampler(random_state=42)
X_train, y_train = ros.fit_resample(X_train, y_train)

print("\nAfter Balancing:")
print("X_train shape:", X_train.shape)
print("y_train distribution:\n", y_train.value_counts())

# ==========================
# 4. READY FOR MODEL TRAINING
# ==========================
# Example with Random Forest (you can change model)

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)

model.fit(X_train, y_train)

# ==========================
# 5. MODEL EVALUATION
# ==========================
y_pred = model.predict(X_test)

print("\nModel Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

### USE BAGGING ON LOGISTIC REGRESSION MODEL

```python
# ==========================
# STEP 1: Import Libraries
# ==========================
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import accuracy_score, classification_report

# ==========================
# STEP 2: Create Base Model
# ==========================
log_reg1 = LogisticRegression(max_iter=1000)

# ==========================
# STEP 3: Create Bagging Model (FIXED PARAMETER)
# ==========================
bagg = BaggingClassifier(
    estimator=log_reg1,   #  FIX: base_estimator → estimator
    n_estimators=50,
    random_state=42,
    n_jobs=-1
)

# ==========================
# STEP 4: Train Model
# ==========================
bagg.fit(X_train, y_train)

# ==========================
# STEP 5: Prediction
# ==========================
bagg_predict = bagg.predict(X_test)

# ==========================
# STEP 6: Evaluation
# ==========================
print("Bagging Accuracy:", accuracy_score(y_test, bagg_predict))
print("\nClassification Report:\n", classification_report(y_test, bagg_predict))
```

### Evaluation

```python
log_bagg_accuracy = accuracy_score(bagg_predict,y_test)
print("Testing accuracy of Logistic regession model after bagging",log_bagg_accuracy*100)
print("Logistic regression testing Classification report: \n",classification_report(bagg_predict,y_test))
```

```python
from flask import Flask, request, jsonify
import pickle
import numpy as np

app = Flask(__name__)

# Load the model safely
try:
    with open('transaction_model.pkl', 'rb') as file:
        model = pickle.load(file)
    print("MODEL LOADED SUCCESSFULLY")
except Exception as e:
    print("MODEL LOADING FAILED:", e)
    model = None

@app.route("/")
def home():
    return "FLASK IS RUNNING"

@app.route("/predict", methods=["POST"])
def predict():
    if model is None:
        return {"error": "Model not loaded"}, 500
    
    data = request.get_json()
    features = np.array(data["features"]).reshape(1, -1)
    prediction = model.predict(features).tolist()
    return jsonify({"prediction": prediction})

if __name__ == "__main__":
    app.run(host="0.0.0.0", port=5000, debug=True, use_reloader=False)
```

## Customer Transaction Prediction Using Machine Learning
### Introduction

In today’s digital era, online transactions have increased rapidly, leading to a significant rise in fraudulent and risky transactions. Predicting customer transactions accurately is essential for financial institutions to minimize losses and enhance customer trust. Machine Learning techniques provide effective solutions by identifying hidden patterns in transaction data.

This project focuses on building and comparing multiple machine learning models to predict customer transactions and selecting the best-performing model based on evaluation metrics.
### Problem Statement

The objective of this project is to develop a predictive model that can classify customer transactions into two categories:

0 – Normal Transaction

1 – Fraudulent / Risky Transaction

The challenge lies in handling numerical and categorical data, scaling features, and selecting the most accurate and reliable model.

### Dataset Description

The dataset consists of customer transaction-related attributes such as:

Transaction Amount

Old Balance

New Balance

Transaction Type

Age

Gender

Payment Mode

Target Variable (Transaction Outcome)

The dataset is structured, labeled, and suitable for supervised machine learning.

### Basic Data Checks

Initial data inspection was performed to ensure data quality.
The following checks were conducted:

Checking dataset shape and structure

Identifying missing values

Detecting duplicate records

Reviewing data types of features

Statistical summary of numerical features

Observation:
The dataset was clean with minimal missing values, which were handled appropriately.

5. Exploratory Data Analysis (EDA)

EDA was carried out to understand data distribution and feature behavior.

Key steps included:

Target variable distribution analysis

Analysis of numerical feature spread

Identification of influential transaction-related features

Checking class imbalance

Insights from EDA:

The target variable showed slight imbalance

Transaction amount and balance features played a crucial role

Some features showed non-linear relationships with the target

6. Data Preprocessing

To prepare the dataset for modeling, the following preprocessing steps were applied:

6.1 Feature Selection

The target variable was separated from input features.

Irrelevant columns were removed.

6.2 Train-Test Split

The dataset was split into 80% training data and 20% testing data.

Stratified sampling was used to maintain class balance.

6.3 Feature Scaling

StandardScaler was applied to normalize numerical features.

Scaling was necessary for algorithms sensitive to feature magnitude.

7. Model Building

Four machine learning models were implemented and trained:

7.1 Logistic Regression

A baseline model used for binary classification.
It assumes a linear relationship between features and the target.

7.2 Random Forest Classifier

An ensemble learning technique that combines multiple decision trees to improve prediction accuracy and reduce overfitting.

7.3 Artificial Neural Network (ANN)

A deep learning model with multiple hidden layers capable of capturing complex non-linear patterns in data.

7.4 XGBoost Classifier

An advanced gradient boosting algorithm known for high performance on structured data.
It efficiently handles feature interactions and minimizes prediction errors.

8. Model Evaluation

Models were evaluated using the following metrics:

Accuracy

Confusion Matrix

Precision, Recall, and F1-Score

Each model was tested on unseen test data to ensure generalization.
